```python
# importing libraries needed
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# reading data set into a dataframe
df = pd.read_csv('https://raw.githubusercontent.com/TrainingByPackt/Data-Wrangling-with-Python/master/Lesson04/Activity06/adult_income_data.csv')
df.head()
```

```python
# creating list of header names for the adult_income_data dataset
names = []
with open('adult.names.csv','r') as f:
    for line in f:
        f.readline()
        var=line.split(":")[0]
        names.append(var)

# view new list of headers
names
```

```python
# removing " from names
import re
names = [item.replace('"', '') for item in names]
names
```

```python
# adding new colum for 'Income'
names.append('Income')

# adding the new list of headers ('names') to the income data set
df = pd.read_csv('https://raw.githubusercontent.com/TrainingByPackt/Data-Wrangling-with-Python/master/Lesson04/Activity06/adult_income_data.csv',names=names)
df.head()
```

```python
# finding missing vaules
df.isnull().sum()
```

```python
# creating a df with only age, education, and occupation
df_subset = df[['age','education','occupation']]
df_subset.head()
```

```python
# histogram of age with bin size of 20
df_subset['age'].hist(bins=20, color = "powderblue", ec="cadetblue")
```

```python
# creating function that strips whitespace characters
def strip_whitespace(s):
    return s.strip()
```

```python
# using the 'apply' method to apply whitespace function to all columns with strings, create a new colum, copy the vaules from old to new, and drop the new column
# education
df_subset['education_stripped']=df['education'].apply(strip_whitespace)
df_subset['education']=df_subset['education_stripped']
df_subset.drop(labels=['education_stripped'],axis=1,inplace=True)
# occupation
df_subset['occupation_stripped']=df['occupation'].apply(strip_whitespace)
df_subset['occupation']=df_subset['occupation_stripped']
df_subset.drop(labels=['occupation_stripped'],axis=1,inplace=True)
# ignoring warning messages
```

```python
# finding the number of people aged between 30 and 50
df_filteredbyage = df_subset[(df_subset['age']>=30) & (df_subset['age']<=50)]
df_filteredbyage.head()
```

```python
# grouping records by age, education, and mean age
df_filteredbyage.groupby(['age','education']).mean()
```

```python
# grouping by occupation withsummary statistics of age
df_filteredbyage.groupby('occupation').describe()['age']
```

```python
# finding which profession has the oldest workers on averge
meanageby_occupation = df_filteredbyage.groupby('occupation').mean()['age']
meanageby_occupationdf = pd.DataFrame(meanageby_occupation)
print(meanageby_occupationdf[meanageby_occupationdf.age == meanageby_occupationdf.age.max()])
```

```python
# finding which profession has the largest share of workers aged above the 75th percentile
# Priv-house-serv
```

```python
df_subset.boxplot(column='age',by='occupation',figsize=(15,6))
plt.xticks(fontsize=15,
        rotation=65)
plt.xlabel("occupation",fontsize=20)
plt.show()
```

```python
# Using a subset to groupby and find outliers in the number of data points per occupation
# Outlires can drive other general statistics higher or lower so it is good to identify these points
occupationstats = df_subset.groupby('occupation').describe()['age']
# plotting the number of data points per occupation
plt.figure(figsize=(15,8))
plt.barh(y=occupationstats.index,
         width=occupationstats['count'],
         color = "powderblue",
         ec="cadetblue")
plt.xlabel("Count",fontsize=12)
plt.ylabel("Occupation",fontsize=12)
```

```python
# merging data using common kyes
# creating first random data set
df1 = df[['age','workclass','occupation']].sample(5,random_state=98)
df1.head()
```

```python
# creating second random data set
df2 = df[['age','workclass','occupation']].sample(5,random_state=56)
df2.head()
```

```python
# merging df1 & df2 using the 'occupation' key
dfmerged = pd.merge(df1,df2,on='occupation',how='inner').drop_duplicates()
dfmerged
```

## Create a series and practice basic arithmetic steps
a. Series 1 = 7.3, -2.5, 3.4, 1.5
i. Index = ‘a’, ‘c’, ‘d’, ‘e’
b. Series 2 = -2.1, 3.6, -1.5, 4, 3.1
i. Index = ‘a’, ‘c’, ‘e’, ‘f’, ‘g’
c. Add Series 1 and Series 2 together and print the results
d. Subtract Series 1 from Series 2 and print the results

```python
# Creating Series 1
data1 = [7.3, -2.5, 3.4, 1.5]
index1 = ['a', 'c', 'd', 'e']
series1 = pd.Series(data = data1, index = index1)
print(series1)
```

```python
# Creating Series 2
data2 = [-2.1, 3.6, -1.5, 4, 3.1]
index2 = ['a', 'c', 'e', 'f','g']
series2 = pd.Series(data = data2, index = index2)
print(series2)
```

```python
# adding series 1 & 2 together
result = series1.add(series2, fill_value=0)
print(result)
```

```python
# subtracting series 1 from 2
result2 = series2 - series1
print(result2)
```
